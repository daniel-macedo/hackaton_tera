```python
import pandas as pd
import simplejson
from tqdm import tqdm

data = []
with open('fake-news-tera.json') as f:
    for i in range(100000):
        line = next(f)
        data.append(simplejson.loads(line))

df = pd.DataFrame(data)
```

```python
pd.options.display.max_columns = len(df.columns)
df.head()
```

```python
import re
import requests
url = '(?:https?):\/\/[\n\S]+'

df_not_relatedtweet=df[df.related_tweet_content.isnull()].copy()
df_not_relatedtweet['content_URL'] = [''.join(re.findall(url, i)) for i in df_not_relatedtweet['content']]
df_not_relatedtweet['content_URL'].head()
```

```python
a = df['content_URL'].unique()
print(len(a))
len(df[df['content_URL']!=''])
```

```python
from tqdm import tqdm

session = requests.Session()
l=[]
for row in tqdm(df['content_URL'], total=len(df['content_URL'])):
    try:
        a = session.head(row, allow_redirects=True).url
    except:
        a =''
        
    l.append(a)
df['content_URL2'] = l
```
